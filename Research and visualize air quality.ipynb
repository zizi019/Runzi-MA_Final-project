```python
import pandas as pd
import folium
from folium import FeatureGroup, LayerControl
from folium.plugins import MarkerCluster
from scipy import stats

df = pd.read_csv("NO2_data_students_filtered.csv")

# Clean data
df["NO2"] = pd.to_numeric(df["NO2"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df.dropna(subset=["Latitude", "Longitude", "NO2", "Year"])
df = df[(df["Latitude"].between(50, 52)) & (df["Longitude"].between(-3, -1))]

# Local anomaly detection
df["lat_bin"] = (df["Latitude"] * 100).astype(int)
df["lon_bin"] = (df["Longitude"] * 100).astype(int)
df["is_outlier"] = False

for (year, lat_bin, lon_bin), group in df.groupby(["Year", "lat_bin", "lon_bin"]):
    if len(group) >= 3:
        z_scores = stats.zscore(group["NO2"])
        outliers = abs(z_scores) > 2.5
        df.loc[group.index[outliers], "is_outlier"] = True

m = folium.Map(location=[51.4545, -2.5879], zoom_start=12)

# Add the year layer
for year in sorted(df["Year"].unique()):
    fg = FeatureGroup(name=f"Year {int(year)}")
    marker_cluster = MarkerCluster().add_to(fg)
    year_df = df[df["Year"] == year]

    for _, row in year_df.iterrows():
        color = "red" if row["is_outlier"] else ("blue" if row["Source"] == "gov" else "green")
        popup = f"""
        <b>Year:</b> {int(row['Year'])}<br>
        <b>NOâ‚‚:</b> {round(row['NO2'], 2)} Âµg/mÂ³<br>
        <b>Source:</b> {row['Source']}<br>
        <b>Status:</b> {"Outlier" if row['is_outlier'] else "Normal"}
        """
        folium.CircleMarker(
            location=(row["Latitude"], row["Longitude"]),
            radius=6 if row["is_outlier"] else 5,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=popup
        ).add_to(marker_cluster)

    fg.add_to(m)

# Only display the layer with outliers
fg_outliers = FeatureGroup(name="ðŸ”´ All Outliers Only")
marker_cluster_outliers = MarkerCluster().add_to(fg_outliers)

for _, row in df[df["is_outlier"]].iterrows():
    popup = f"""
    <b>Year:</b> {int(row['Year'])}<br>
    <b>NOâ‚‚:</b> {round(row['NO2'], 2)} Âµg/mÂ³<br>
    <b>Source:</b> {row['Source']}<br>
    <b>Status:</b> Outlier
    """
    folium.CircleMarker(
        location=(row["Latitude"], row["Longitude"]),
        radius=7,
        color="red",
        fill=True,
        fill_color="red",
        fill_opacity=0.9,
        popup=popup
    ).add_to(marker_cluster_outliers)

fg_outliers.add_to(m)

LayerControl().add_to(m)

# Save map
m.save("NO2_map_years_with_outlier_layer.html")
print("The map has been generated, including a year layer and a separate outlier layer!")
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("NO2_data_students_filtered.csv")

# Make sure the numerical fields are correct
df["NO2"] = pd.to_numeric(df["NO2"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df.dropna(subset=["Latitude", "Longitude", "NO2", "Year"])

# Geographical location filtering
df = df[(df["Latitude"].between(50, 52)) & (df["Longitude"].between(-3, -1))]

# Add an outlier marker
df["is_outlier"] = False

for year in df["Year"].unique():
    subset = df[df["Year"] == year]
    z_scores = stats.zscore(subset["NO2"])
    outliers = abs(z_scores) > 2.5 
    df.loc[subset.index[outliers], "is_outlier"] = True

# Statistical results
total_points = len(df)
outlier_points = df["is_outlier"].sum()
percentage = round((outlier_points / total_points) * 100, 2)

print(f"The number of outliers:{outlier_points}")
print(f"Total number of data points:{total_points}")
print(f"Proportion of outliers:{percentage}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv("NO2_data_students_filtered.csv")

# Type conversion and cleaning
df["NO2"] = pd.to_numeric(df["NO2"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df.dropna(subset=["NO2", "Year"])

grouped = df.groupby(["Year", "Source"])["NO2"].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
for source in grouped["Source"].unique():
    subset = grouped[grouped["Source"] == source]
    plt.plot(subset["Year"], subset["NO2"], marker='o', label=source)

plt.title("Average NOâ‚‚ Concentration by Year (Student vs Gov)")
plt.xlabel("Year")
plt.ylabel("NOâ‚‚ (Âµg/mÂ³)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("NO2_trend_by_year.png", dpi=300)
plt.show()

gov_df = df[df["Source"] == "gov"]

pre_caz = gov_df[gov_df["Year"] <= 2021]["NO2"]
post_caz = gov_df[gov_df["Year"] >= 2022]["NO2"]

t_stat, p_value = stats.ttest_ind(pre_caz.dropna(), post_caz.dropna(), equal_var=False)

print(f"T-test resultï¼št = {t_stat:.2f}, p = {p_value:.4f}")
if p_value < 0.05:
    print("The differences before and after the policy are statistically significantï¼ˆp < 0.05ï¼‰")
else:
    print("There is no significant difference before and after the policy")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load dataset
df = pd.read_csv("NO2_data_students_filtered.csv")

# Data cleaning
df["NO2"] = pd.to_numeric(df["NO2"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df.dropna(subset=["NO2", "Latitude", "Longitude", "Year"])

# Separate student and government data
df_students = df[df["Source"] == "student"]
df_gov = df[df["Source"] == "gov"]

paired_data = []

# Match data by year
for year in sorted(df_students["Year"].unique()):
    students_year = df_students[df_students["Year"] == year]
    gov_year = df_gov[df_gov["Year"] == year]

    for _, student_row in students_year.iterrows():
        # Find the nearest government point (within 0.01Â° radius)
        nearby_gov = gov_year[
            (abs(gov_year["Latitude"] - student_row["Latitude"]) <= 0.01) &
            (abs(gov_year["Longitude"] - student_row["Longitude"]) <= 0.01)
        ]
        if not nearby_gov.empty:
            gov_row = nearby_gov.iloc[0]
            paired_data.append({
                "Year": year,
                "Student_NO2": student_row["NO2"],
                "Gov_NO2": gov_row["NO2"]
            })

df_paired = pd.DataFrame(paired_data)

# Calculate correlation and plot
if not df_paired.empty:
    r, p = pearsonr(df_paired["Gov_NO2"], df_paired["Student_NO2"])
    print(f"Pearson correlation: r = {r:.2f}, p = {p:.4f}")

    plt.figure(figsize=(8, 6))

    # Scatter plot (blue points for paired measurements)
    plt.scatter(df_paired["Gov_NO2"], df_paired["Student_NO2"],
                alpha=0.6, label="Paired measurements (Student vs Gov)")

    # 1:1 reference line (red dashed line)
    plt.plot([0, df_paired["Gov_NO2"].max()],
             [0, df_paired["Gov_NO2"].max()],
             color="red", linestyle="--", label="1:1 line")

    # Axis labels and title
    plt.xlabel("Government NOâ‚‚ (Âµg/mÂ³)")
    plt.ylabel("Student NOâ‚‚ (Âµg/mÂ³)")
    plt.title("Student vs Government NOâ‚‚ Measurements")
    plt.grid(True)

    # Show Pearson r and p-value in the bottom-right corner of the plot
    plt.text(0.95, 0.05,
             f"r = {r:.2f}\np = {p:.4f}",
             transform=plt.gca().transAxes,
             fontsize=10,
             ha="right", va="bottom",
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.6))

    # Add legend
    plt.legend()

    # Save and show plot
    plt.tight_layout()
    plt.savefig("Student_vs_Gov_NO2.png", dpi=300)
    plt.show()

else:
    print("No paired data found within matching range.")
```

```python
import pandas as pd
import folium
from folium import LayerControl
from folium.plugins import HeatMap

df = pd.read_csv("NO2_data_students_filtered.csv")

# Clean
df["NO2"] = pd.to_numeric(df["NO2"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df.dropna(subset=["Latitude", "Longitude", "NO2", "Year"])
df = df[(df["Latitude"].between(50, 52)) & (df["Longitude"].between(-3, -1))]

# Create basic map
m = folium.Map(location=[51.4545, -2.5879], zoom_start=12)

years = sorted(df["Year"].unique())

for year in years:
    fg = folium.FeatureGroup(name=f"Year {int(year)}")
    year_data = df[df["Year"] == year]
    max_no2 = year_data["NO2"].quantile(0.95)
    heat_data = year_data[["Latitude", "Longitude", "NO2"]].values.tolist()

    if heat_data:
        HeatMap(
            data=heat_data,
            radius=10,
            blur=8,
            min_opacity=0.4,
            max_val=max_no2,
            gradient={
                0.2: 'blue',
                0.4: 'lime',
                0.6: 'yellow',
                0.8: 'orange',
                1.0: 'red'
            }
        ).add_to(fg)

        fg.add_to(m)

LayerControl().add_to(m)

# Save map
m.save("NO2_Heatmap_Optimized.html")
print("The optimized heat map has been saved as NO2_Heatmap_Optimized.html")
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

df = pd.read_csv("merged_NO2_all_sources.csv")
df = df.dropna(subset=["Latitude", "Longitude", "NO2", "Year"])

# Build GeoDataFrame
df["geometry"] = df.apply(lambda row: Point(row["Longitude"], row["Latitude"]), axis=1)
df_gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")

caz = gpd.read_file("Clean_Air_Zone.geojson")
caz = caz.to_crs(epsg=4326)

df_gdf["In_CAZ"] = df_gdf.geometry.within(caz.unary_union)
df_gdf["Post_CAZ"] = df_gdf["Year"] >= 2023
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the mean
avg = df_gdf.groupby(["Year", "In_CAZ"])["NO2"].mean().reset_index()
avg["Zone"] = avg["In_CAZ"].map({True: "Inside CAZ", False: "Outside CAZ"})

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg, x="Year", y="NO2", hue="Zone", marker="o")
plt.title("Mean NOâ‚‚ Levels Inside vs Outside CAZ (by Year)")
plt.ylabel("NOâ‚‚ (Âµg/mÂ³)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

pre = df_gdf[(df_gdf["In_CAZ"]) & (df_gdf["Year"] < 2023)]["NO2"]
post = df_gdf[(df_gdf["In_CAZ"]) & (df_gdf["Year"] >= 2023)]["NO2"]

t_stat, p_val = ttest_ind(pre, post, equal_var=False)
print(f"Inside CAZ: t = {t_stat:.2f}, p = {p_val:.4f}")
df_gdf.to_csv("NO2_data_with_CAZ_flag.csv", index=False)
```

```python
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
from folium import FeatureGroup, LayerControl

df = pd.read_csv("NO2_data_students_filtered.csv")
caz = gpd.read_file("Clean_Air_Zone.geojson")

# Clean data
df["NO2"] = pd.to_numeric(df["NO2"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df = df.dropna(subset=["Latitude", "Longitude", "NO2", "Year"])
df = df[(df["Latitude"].between(50, 52)) & (df["Longitude"].between(-3, -1))]

# Initialize the map
m = folium.Map(location=[51.4545, -2.5879], zoom_start=12)

# Add the CAZ area
caz_clean = caz[["geometry"]].copy()
folium.GeoJson(
    caz_clean,
    name="Clean Air Zone",
    style_function=lambda feature: {
        "fillColor": "#ff7800",
        "color": "#ff0000",
        "weight": 2,
        "fillOpacity": 0.1
    }
).add_to(m)

# Add marker layers by year
years = sorted(df["Year"].dropna().unique())

for year in years:
    fg = FeatureGroup(name=str(int(year)))
    year_df = df[df["Year"] == year]
    marker_cluster = MarkerCluster().add_to(fg)

    for _, row in year_df.iterrows():
        popup = f"""
        <b>Year:</b> {int(row['Year'])}<br>
        <b>NOâ‚‚:</b> {round(row['NO2'], 2)} Âµg/mÂ³<br>
        <b>Source:</b> {row['Source']}
        """
        folium.CircleMarker(
            location=(row["Latitude"], row["Longitude"]),
            radius=5,
            color="blue" if row["Source"] == "gov" else "green",
            fill=True,
            fill_color="blue" if row["Source"] == "gov" else "green",
            fill_opacity=0.7,
            popup=popup
        ).add_to(marker_cluster)

    fg.add_to(m)

# Add layer controller
LayerControl().add_to(m)
m.save("NO2_map_with_CAZ_by_year.html")
print("The map has been saved as NO2_map_with_CAZ_by_year.html")
```
